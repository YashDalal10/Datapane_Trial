```python
import yfinance as yf
```

```python
today = date.today()
today
```

```python
yesterday = date.today() - timedelta(1)
yesterday
```

```python
year = date.today() - timedelta(365)
year
```

```python
tickers = ['BEL.NS','CIPLA.NS','GAIL.NS','HAL.NS','HDFCBANK.NS','HDFC.NS','HDFCLIFE.NS','HFCL.NS','ICICIBANK.NS','IOC.NS','IRCTC.NS','INFY.NS','KOTAKBANK.NS','LT.NS','LICI.NS',
           'M&M.NS','MIDHANI.NS','RELIANCE.NS','SBIN.NS','SUNPHARMA.NS','TCS.NS','TATAMTRDVR.NS','TATAMOTORS.NS','TATAPOWER.NS','TATASTEEL.NS','WIPRO.NS','YESBANK.NS']

tickers
```

```python
# hdfc_bk = yf.Ticker('HDFCBANK.NS')
# hdfc_bk_data = hdfc_bk.history(interval = '1d', start = yday, end = today)['Close'][0]
# hdfc_bk_data
```

```python
data = pd.DataFrame()
for i in range(len(tickers)):
  tcker = yf.Ticker(tickers[i])
  new_data = tcker.history(interval = '1d', start = year, end = today)
  new_data['ticker'] = tickers[i]
  data = pd.concat([data, new_data])

data.head(10)
# hdfc_bk = yf.Ticker('HDFCBANK.NS')
# hdfc_bk_data = hdfc_bk.history(interval = '1d', start = yday, end = today)
# hdfc_bk_data
```

```python
data_copy = data.copy()
data_copy
```

```python
data_copy['Date'] = data_copy.index
data_copy.sample(10)
```

```python
data_copy = data_copy.reset_index(drop=True)
```

```python
data_copy['Date'] = pd.to_datetime(data_copy['Date'])
data_copy['Date'] = data_copy['Date'].dt.date
data_copy
```

```python
latest_df1 = data_copy[data_copy['Date']==yesterday]
latest_df1
```
